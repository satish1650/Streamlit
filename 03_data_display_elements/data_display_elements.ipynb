✅ `st.dataframe()`

`st.dataframe()` is used to **display interactive tables** (like Excel or Google Sheets) in your `Streamlit` app. It's ideal for viewing **Pandas DataFrame**s, **lists of lists**, or **dictionaries** in a tabular format.

```Python
🔸Example

import streamlit as st
import pandas as pd

data= pd.DataFrame(
    {
        "Name": ["Alice", "Bob", "Charlie"],
         "Age": [25, 30, 35],
         "Score": [88, 92, 95]
    }
)

st.dataframe(data)

#### Output



🔸Example with `use_container_width`:

```Python

# This makes it responsive and fills the page width.
st.dataframe(data, use_container_width=True)

#### Output



#### 🔁 Dynamic Styling with Pandas

```Python
# This highlights the max value in each column.
styled= data.style.highlight_max(axis=0)
st.dataframe(styled)

#### Output
