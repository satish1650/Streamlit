### `st.chat_input()`

**st.chat_input()** in **Streamlit** is the widget that lets you create a **chat-style text input box** at the bottom of your app (similar to how chat apps look).


<hr>

#### `Example-1`
```Python
import streamlit as st

st.chat_input(placeholder="Please enter input here...")
```
<hr>

#### `Streamlit Output`


<hr>

#### `Example-2`
```Python
user_input = st.chat_input(placeholder="Please enter input here...")
if user_input:
    print(st.write(f"Your input is {user_input}"))
```
<hr>

#### `Streamlit Output`

#### `Example-3`
```Python
user_input = st.chat_input(placeholder="Please enter input here...")
if user_input:
    print(st.write(f"Your input is \n", {user_input}))
```

#### `Example-4`
```Python
if user_input := st.chat_input(placeholder="Please enter input here..."):
    print(st.write(f"Your input is \n", {user_input}))
```
