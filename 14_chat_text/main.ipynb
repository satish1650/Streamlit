### `st.text_input()`
**st.text_input()** is a basic **single-line text box** where a user can type something.
It’s one of Streamlit’s core input widgets.

<hr>

#### `Example-1`
```Python
import streamlit as st

# Basic usage
name = st.text_input("Enter your name:")

# Use entered value
if name:
    st.write(f"Hello, {name}! 👋")
```

#### `Example-2`
```Python
import streamlit as st

# Basic usage
name = st.text_input("Enter your name:")

# Use entered value
if name:
    st.write(f"Hello, {name}! 👋")

# With a placeholder and password type
password = st.text_input("Enter password:", type="password", placeholder="••••••")
```
