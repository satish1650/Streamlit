#### `st.markdown()`
<hr>

The `st.markdown()` function in **Streamlit** lets you write and display Markdown-formatted text in your app - including headings, lists, bold/italic text, links, images, code blocks, and even some HTML/CSS hacks.

<hr>

#### `Example-1`
```Python
import streamlit as st

# Bold and italic text
st.markdown("Your *markdown* **content** here.")

st.markdown("This is **bold** and this is *italic*.")

st.markdown("This is **BOLD** and this is ***ITALIC***.")

st.markdown("You can also use `inline code` in your text.")
```
<hr>

#### `Streamlit Output`



<hr>

#### `Example-2`
```Python
import streamlit as st

st.divider()

# Headings & Text Formatting
st.markdown("# H1 Heading")
st.markdown("## H2 Heading")
st.markdown("### H3 Heading")

st.markdown("**Bold Text** and *Italic Text*")

st.divider()
```

#### `Example-3`
```Python
import streamlit as st

st.divider()

# List 
st.markdown("""
- Bullet 1
- Bullet 2
  - Sub-bullet

1. Numbered item
2. Another item
""")

st.divider()
```

#### `Example-4`
```Python
import streamlit as st

st.divider()

# Links & Images
st.markdown("[Visit Google](https://www.google.com)")

st.markdown("![Streamlit Logo](https://streamlit.io/images/brand/streamlit-logo-secondary-colormark-darktext.png)")

st.divider()
```

#### `Example-5`
```Python
import streamlit as st

st.divider()

# Inline Code & Code Blocks
st.markdown("Use `print()` to display output in Python.")

st.markdown("""
```python
def greet():
    print("Hello, Streamlit!")
""")

st.divider()
```
